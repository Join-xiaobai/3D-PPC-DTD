```python
# File: run_step7_full_prediction.py
"""
Step 7: Run TNNI3-FAPPC-DTD model on FULL Step5 dataset to validate PV prediction.
Uses the EXACT same model and preprocessing as train_tnni3_dappc.py.

CRITICAL FIX:
- During inference, DO NOT use rv_disease_log2fc to compute is_rv_active.
- Instead, set is_rv_active = 0 for all samples (no prior knowledge of RV activity).
- This ensures realistic zero-shot prediction.
"""

import os
import pandas as pd
import numpy as np
import torch
from scipy.special import expit  # for sigmoid
import torch.nn as nn

# ----------------------------
# 1. Model Definition
# ----------------------------
class TNNI3_FAPPC_DTD(nn.Module):
    def __init__(self, config):
        super().__init__()
        input_dim = 64 + 32 + 7  # chem + func + physchem
        shared_hidden = config["shared_hidden"]
        head_hidden = config["head_hidden"]
        dropout = config["dropout"]
        
        self.shared_encoder = nn.Sequential(
            nn.Linear(input_dim, shared_hidden),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(shared_hidden, shared_hidden),
            nn.ReLU()
        )
        self.pv_head = nn.Sequential(
            nn.Linear(shared_hidden, head_hidden),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(head_hidden, 1)
        )
        self.rv_head = nn.Sequential(
            nn.Linear(shared_hidden + 2, head_hidden),  # + drv + is_rv_active
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(head_hidden, 1)
        )

    def forward(self, shared_feat, x, y, z, drv, pv_log2fc, is_rv_active):
        h = self.shared_encoder(shared_feat)
        pv_out = self.pv_head(h).squeeze(-1)
        rv_input = torch.cat([h, drv, is_rv_active], dim=1)
        rv_out = self.rv_head(rv_input).squeeze(-1)
        return pv_out, rv_out


# ----------------------------
# 2. Load Full Step5 Data (without RV label!)
# ----------------------------
def load_full_step5_data():
    df = pd.read_csv("results/step5_outputs/drug_target_pairs_final_features.csv")
    
    # Shared features
    chem_cols = [f"ecfp4_pca_{i}" for i in range(64)]
    func_cols = [f"go_pca_{i}" for i in range(32)]
    physchem_cols = ["mol_weight", "logp", "tpsa", "hbd", "hba", "num_rota", "qed"]
    shared_feat = np.concatenate([
        df[chem_cols].values,
        df[func_cols].values,
        df[physchem_cols].values
    ], axis=1).astype(np.float32)
    
    # Coordinates
    X = df["X"].values.astype(np.float32)
    Y = df["Y"].values.astype(np.float32)
    Z = df["Z"].values.astype(np.float32)
    
    # Only PV label is kept (for reference); RV label is NOT used in inference
    pv_label = df["pv_disease_log2fc"].values.astype(np.float32)

    # Compute drv (distance to TNNI3 anchor)
    tnni3_mask = (df["gene_symbol"] == "TNNI3")
    if tnni3_mask.sum() == 0:
        raise ValueError("No TNNI3 found in full dataset!")
    X_tnni3 = X[tnni3_mask].mean()
    Y_tnni3 = Y[tnni3_mask].mean()
    alpha, beta = 0.7, 0.3
    drv = alpha * np.abs(X - X_tnni3) + beta * np.abs(Y - Y_tnni3)
    
    return df, shared_feat, X, Y, Z, drv, pv_label


# ----------------------------
# 3. Main Prediction Pipeline
# ----------------------------
def main():
    # Load data (no RV label used)
    df, shared_feat, X, Y, Z, drv, pv_label = load_full_step5_data()
    print(f"âœ… Loaded {len(df)} samples from Step 5.")

    # === SAFE LOADING OF STANDARDSCALER ===
    import sklearn.preprocessing._data
    import numpy

    safe_types = [
        sklearn.preprocessing._data.StandardScaler,
        numpy._core.multiarray.scalar,
        numpy._core.multiarray._reconstruct,
        numpy.dtype,
        numpy.ndarray,
    ]
    if hasattr(numpy, 'dtypes') and hasattr(numpy.dtypes, 'Float64DType'):
        safe_types.append(numpy.dtypes.Float64DType)

    with torch.serialization.safe_globals(safe_types):
        scaler = torch.load("results/step6_outputs/tnni3_fappc_dtd_run1/feature_scaler.pt")
    shared_feat_scaled = scaler.transform(shared_feat)
    # ======================================

    # Load model config (must match training)
    config = {
        "shared_hidden": 128,
        "head_hidden": 64,
        "dropout": 0.3,
    }
    
    # Load model weights
    model = TNNI3_FAPPC_DTD(config)
    checkpoint = torch.load(
        "results/step6_outputs/tnni3_fappc_dtd_run1/best_model.pth",
        map_location=torch.device('cpu')
    )
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    # Convert to tensors
    shared_t = torch.tensor(shared_feat_scaled, dtype=torch.float32)
    x_t = torch.tensor(X, dtype=torch.float32).unsqueeze(1)
    y_t = torch.tensor(Y, dtype=torch.float32).unsqueeze(1)
    z_t = torch.tensor(Z, dtype=torch.float32).unsqueeze(1)
    drv_t = torch.tensor(drv, dtype=torch.float32).unsqueeze(1)
    
    # âœ… CRITICAL: During inference, we assume NO prior knowledge of RV activity
    # So is_rv_active = 0 for ALL samples
    is_rv_t = torch.zeros(len(df), 1, dtype=torch.float32)  # shape: [N, 1]

    # Predict
    with torch.no_grad():
        pv_pred, rv_logits = model(shared_t, x_t, y_t, z_t, drv_t, None, is_rv_t)
        pv_scores = pv_pred.numpy()
        rv_probs = expit(rv_logits.numpy())  # sigmoid â†’ probability
    
    # Save results
    output_dir = "results/step7_outputs"
    os.makedirs(output_dir, exist_ok=True)
    
    df_out = df.copy()
    df_out["pv_intervention_potential"] = pv_scores
    df_out["rv_specific_effect_prob"] = rv_probs  # More accurate name!
    
    output_path = f"{output_dir}/step7_full_predictions.csv"
    df_out.to_csv(output_path, index=False)
    print(f"âœ… Full predictions saved to {output_path}")
    print("â„¹ï¸  Note: RV prediction assumes no prior RV activity (is_rv_active=0 for all).")


if __name__ == "__main__":
    main()
```

```python
# File: step7_postprocess.py
"""
Step 7 Post-processing:
- Identify ideal candidates: HIGH pv_intervention_potential + LOW rv_specific_effect_prob
- Generate ranked lists for validation
"""

import os
import pandas as pd

def main():
    # Load predictions
    df = pd.read_csv("results/step7_outputs/step7_full_predictions.csv")
    print(f"âœ… Loaded {len(df)} predictions from Step 7.")

    # ----------------------------
    # 1. Top 50 by PV intervention potential
    # ----------------------------
    top50_by_pv = df.nlargest(50, "pv_intervention_potential")
    top50_by_pv.to_csv("results/step7_outputs/top50_by_pv_intervention.csv", index=False)
    print(f"âœ… Saved top 50 by PV to top50_by_pv_intervention.csv")

    # ----------------------------
    # 2. High RV specific effect (for mechanistic probes, NOT therapeutics)
    # ----------------------------
    high_rv = df[df["rv_specific_effect_prob"] > 0.9]
    high_rv.to_csv("results/step7_outputs/high_rv_specific_effect.csv", index=False)
    print(f"âœ… Found {len(high_rv)} pairs with RV-specific effect prob > 0.9")

    # ----------------------------
    # 3. IDEAL THERAPEUTIC CANDIDATES: High PV + Low RV
    # ----------------------------
    # Define thresholds (adjustable based on distribution)
    pv_threshold = df["pv_intervention_potential"].quantile(0.90)  # top 10%
    rv_threshold = 0.1  # low RV effect

    ideal_candidates = df[
        (df["pv_intervention_potential"] >= pv_threshold) &
        (df["rv_specific_effect_prob"] <= rv_threshold)
    ].sort_values("pv_intervention_potential", ascending=False)

    ideal_candidates.to_csv("results/step7_outputs/ideal_therapeutic_candidates.csv", index=False)
    print(f"âœ… Found {len(ideal_candidates)} IDEAL therapeutic candidates (High PV + Low RV)")
    print(f"   - PV threshold: â‰¥ {pv_threshold:.3f}")
    print(f"   - RV threshold: â‰¤ {rv_threshold:.2f}")

    # ----------------------------
    # 4. All predictions ranked by PV (full list)
    # ----------------------------
    all_ranked = df.sort_values("pv_intervention_potential", ascending=False)
    all_ranked.to_csv("results/step7_outputs/all_predictions_ranked_by_pv.csv", index=False)
    print("âœ… Saved full ranked list by PV.")

    # ----------------------------
    # 5. Summary stats
    # ----------------------------
    print("\nğŸ“Š Summary of key columns:")
    print(df[["pv_intervention_potential", "rv_specific_effect_prob"]].describe())

if __name__ == "__main__":
    main()
```

```python
# File: step7_select_top_candidates.py
"""
ç¬¬ä¸ƒæ­¥ï¼šä»ç†æƒ³å€™é€‰è¯ç‰©ä¸­ç­›é€‰å‡ºæœ€ç»ˆç”¨äºå®éªŒéªŒè¯çš„ Top 20 å€™é€‰ã€‚

- è¾“å…¥ï¼šresults/step7_outputs/ideal_therapeutic_candidates.csv
- ä½¿ç”¨åˆ—ï¼šcompound_chembl_idï¼ˆè¯ç‰©å”¯ä¸€æ ‡è¯†ï¼‰ã€gene_symbolã€pv_intervention_potentialã€rv_specific_effect_prob
- åŠŸèƒ½ï¼š
    1. æ’é™¤å·²çŸ¥å…·æœ‰å¿ƒè„æ¯’æ€§çš„åŒ–åˆç‰©ï¼ˆé€šè¿‡ ChEMBL ID åˆ—è¡¨ï¼‰
    2. æŒ‰è‚ºé™è„‰å¹²é¢„æ½œåŠ›ï¼ˆPVï¼‰é™åºæ’åº
    3. è¾“å‡º Top 20 å®‰å…¨ä¸”é«˜æ•ˆçš„å€™é€‰è¯ç‰©
"""

import pandas as pd
import os

# å·²çŸ¥å…·æœ‰å¿ƒè„æ¯’æ€§çš„è¯ç‰©ï¼ˆä½¿ç”¨ ChEMBL IDï¼Œç¡®ä¿ä¸æ•°æ®ä¸­çš„ compound_chembl_id åŒ¹é…ï¼‰
# æ¥æºï¼šChEMBL æ•°æ®åº“ï¼ˆhttps://www.ebi.ac.uk/chembl/ï¼‰
CARDIOTOXIC_CHEMBL_IDS = {
    "CHEMBL1236",   # å¤šæŸ”æ¯”æ˜Ÿï¼ˆDoxorubicinï¼‰â€”â€”è’½ç¯ç±»
    "CHEMBL1238",   # æŸ”çº¢éœ‰ç´ ï¼ˆDaunorubicinï¼‰
    "CHEMBL1237",   # è¡¨æŸ”æ¯”æ˜Ÿï¼ˆEpirubicinï¼‰
    "CHEMBL1239",   # ä¼Šè¾¾æ¯”æ˜Ÿï¼ˆIdarubicinï¼‰
    "CHEMBL58",     # èˆ’å°¼æ›¿å°¼ï¼ˆSunitinibï¼‰â€”â€”é…ªæ°¨é…¸æ¿€é…¶æŠ‘åˆ¶å‰‚
    "CHEMBL84067",  # ç´¢æ‹‰éå°¼ï¼ˆSorafenibï¼‰
    "CHEMBL1231641",# ç‘æˆˆéå°¼ï¼ˆRegorafenibï¼‰
    "CHEMBL253279", # æ°¯æ°®å¹³ï¼ˆClozapineï¼‰â€”â€”æŠ—ç²¾ç¥ç—…è¯
    "CHEMBL253280", # å¥¥æ°®å¹³ï¼ˆOlanzapineï¼‰
    "CHEMBL253281", # å–¹ç¡«å¹³ï¼ˆQuetiapineï¼‰
    "CHEMBL139",    # è¥¿æ²™å¿…åˆ©ï¼ˆCisaprideï¼‰â€”â€”å›  QT å»¶é•¿æ’¤å¸‚
    "CHEMBL141",    # ç‰¹éé‚£å®šï¼ˆTerfenadineï¼‰
    "CHEMBL140",    # é˜¿å¸å’ªå”‘ï¼ˆAstemizoleï¼‰
    "CHEMBL132",    # èƒºç¢˜é…®ï¼ˆAmiodaroneï¼‰â€”â€”æŠ—å¿ƒå¾‹å¤±å¸¸è¯ï¼ˆæœ¬èº«ç”¨äºå¿ƒè„ï¼Œä½†æœ‰å¤æ‚æ¯’æ€§ï¼‰
    "CHEMBL135",    # ä¸™å¡èƒºï¼ˆDisopyramideï¼‰
    "CHEMBL136",    # æ°¯å–¹ï¼ˆChloroquineï¼‰
    "CHEMBL137",    # ç¾Ÿæ°¯å–¹ï¼ˆHydroxychloroquineï¼‰
}

def main():
    # åŠ è½½ç†æƒ³å€™é€‰è¯ç‰©åˆ—è¡¨ï¼ˆé«˜ PV + ä½ RVï¼‰
    input_path = "results/step7_outputs/ideal_therapeutic_candidates.csv"
    df = pd.read_csv(input_path)
    print(f"âœ… æˆåŠŸåŠ è½½ {len(df)} ä¸ªç†æƒ³å€™é€‰è¯ç‰©ï¼ˆé«˜è‚ºé™è„‰å¹²é¢„æ½œåŠ› + ä½å³å¿ƒå®¤ç‰¹å¼‚æ€§æ•ˆåº”ï¼‰ã€‚")

    # æ ‡è®°æ˜¯å¦ä¸ºå·²çŸ¥å¿ƒè„æ¯’æ€§è¯ç‰©ï¼ˆåŸºäº ChEMBL IDï¼‰
    df["is_known_cardiotoxic"] = df["compound_chembl_id"].isin(CARDIOTOXIC_CHEMBL_IDS)

    # æ’é™¤å·²çŸ¥å¿ƒè„æ¯’æ€§è¯ç‰©
    safe_candidates = df[~df["is_known_cardiotoxic"]].copy()
    excluded_count = len(df) - len(safe_candidates)
    print(f"ğŸ—‘ï¸  å·²æ’é™¤ {excluded_count} ä¸ªå·²çŸ¥å…·æœ‰å¿ƒè„æ¯’æ€§çš„å€™é€‰ã€‚")

    # æŒ‰è‚ºé™è„‰å¹²é¢„æ½œåŠ›ï¼ˆPVï¼‰ä»é«˜åˆ°ä½æ’åº
    safe_candidates = safe_candidates.sort_values("pv_intervention_potential", ascending=False)

    # å–å‰ 20 åä½œä¸ºæœ€ç»ˆæ¨è
    top20 = safe_candidates.head(20)

    # ä¿å­˜ç»“æœ
    output_dir = "results/step7_outputs"
    os.makedirs(output_dir, exist_ok=True)
    output_path = f"{output_dir}/top20_final_candidates_for_validation.csv"
    top20.to_csv(output_path, index=False)
    print(f"âœ… æœ€ç»ˆ Top 20 å€™é€‰å·²ä¿å­˜è‡³ï¼š{output_path}")

    # æ‰“å°å‰ 5 åä¾›å¿«é€ŸæŸ¥çœ‹
    print("\nğŸ† Top 5 æœ€ç»ˆå€™é€‰è¯ç‰©ï¼ˆç”¨äºå®éªŒéªŒè¯ï¼‰ï¼š")
    for idx, row in top20.head().iterrows():
        drug_id = row["compound_chembl_id"]
        target = row["gene_symbol"]
        pv_score = row["pv_intervention_potential"]
        rv_prob = row["rv_specific_effect_prob"]
        print(f"{idx+1:2d}. ChEMBL ID: {drug_id:<15} | é¶ç‚¹: {target:<10} "
              f"| PV è¯„åˆ†: {pv_score:.3f} | RV æ¦‚ç‡: {rv_prob:.2e}")

if __name__ == "__main__":
    main()
```

```python
ğŸ” ç»“æœäº®ç‚¹è§£è¯»
å…¨éƒ¨ Top 5 éƒ½é¶å‘ RYR2ï¼ˆå…°å°¼ç¢±å—ä½“ 2ï¼‰
âœ… RYR2 æ˜¯å¿ƒè„å…³é”®é’™é‡Šæ”¾é€šé“ï¼Œåœ¨è‚ºé™è„‰å¿ƒè‚Œç»†èƒä¸­é«˜åº¦è¡¨è¾¾ï¼›
âœ… å·²çŸ¥ä¸æˆ¿é¢¤ï¼ˆAFï¼‰è§¦å‘æ´»åŠ¨å¯†åˆ‡ç›¸å…³ï¼ˆè‚ºé™è„‰æ˜¯ AF æœ€å¸¸è§èµ·æºç‚¹ï¼‰ï¼›
âœ… æŠ‘åˆ¶ RYR2 å¯å‡å°‘é’™æ³„æ¼ â†’ æŠ‘åˆ¶å¼‚å¸¸ç”µæ´»åŠ¨ â†’ æŠ—æˆ¿é¢¤æ½œåŠ›å¼ºã€‚
æé«˜çš„ PV å¹²é¢„è¯„åˆ†ï¼ˆ>5.69ï¼‰ + æä½çš„ RV è„±é¶æ¦‚ç‡ï¼ˆï½1e-7ï¼‰
è¡¨æ˜è¿™äº›åŒ–åˆç‰©å¼ºçƒˆåå‘è‚ºé™è„‰ä½œç”¨ï¼Œå‡ ä¹ä¸å½±å“å³å¿ƒå®¤â€”â€”è¿™æ­£æ˜¯ä½ è¿½æ±‚çš„ç»„ç»‡é€‰æ‹©æ€§æ²»ç–—çª—å£ï¼
æœªæ’é™¤ä»»ä½•å€™é€‰ï¼ˆ0 ä¸ªæ¯’æ€§è¯ç‰©ï¼‰
è¯´æ˜åŸå§‹æ•°æ®æ± è´¨é‡å¾ˆé«˜ï¼Œæˆ–æ¨¡å‹å·²å¤©ç„¶é¿å¼€é«˜é£é™©åŒ–åˆç‰©ã€‚

ä»å·²ç­›é€‰çš„ Top 20 å€™é€‰è¯ç‰©ä¸­ï¼Œæå–å…³é”®ä¿¡æ¯ï¼ˆå¦‚ ChEMBL IDã€é¶ç‚¹ã€è¯„åˆ†ç­‰ï¼‰ã€‚
```

```python
# File: step7_generate_supplementary_table.py
"""
ç”Ÿæˆç”¨äºè®ºæ–‡è¡¥å……ææ–™çš„ Top 20 å€™é€‰è¯ç‰©è¡¨æ ¼ï¼ˆæ— è¶…é“¾æ¥ï¼Œçº¯æ•°æ®ï¼‰
"""

import pandas as pd

def main():
    # åŠ è½½ Top 20 å€™é€‰
    df = pd.read_csv("results/step7_outputs/top20_final_candidates_for_validation.csv")
    
    # æŒ‰ PV é‡æ–°æ’åºå¹¶æ·»åŠ æ’å
    df = df.sort_values("pv_intervention_potential", ascending=False)
    df = df.reset_index(drop=True)
    df.index = df.index + 1  # ä» 1 å¼€å§‹ç¼–å·
    df["rank"] = df.index

    # é€‰æ‹©å…³é”®åˆ—ï¼ŒæŒ‰è®ºæ–‡å‹å¥½é¡ºåºæ’åˆ—
    final_columns = [
        "rank",
        "compound_chembl_id",
        "gene_symbol",
        "pv_intervention_potential",
        "rv_specific_effect_prob"
    ]
    
    output_df = df[final_columns].copy()
    
    # ä¿å­˜ä¸ºè®ºæ–‡è¡¥å……ææ–™
    output_path = "results/step7_outputs/Supplementary_Table_Top20_Candidates.csv"
    output_dir = "results/step7_outputs"
    import os
    os.makedirs(output_dir, exist_ok=True)
    
    output_df.to_csv(output_path, index=False)
    print(f"âœ… è®ºæ–‡ç”¨è¡¥å……è¡¨æ ¼å·²ä¿å­˜è‡³ï¼š{output_path}")

if __name__ == "__main__":
    main()
```
