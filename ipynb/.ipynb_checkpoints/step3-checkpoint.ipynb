```python
import sqlite3
import pandas as pd
import os

# ç¡®ä¿è¾“å‡ºç›®å½•å­˜åœ¨
os.makedirs("data/processed/step3", exist_ok=True)

conn = sqlite3.connect("data/raw/step1/chembl_36/chembl_36_sqlite/chembl_36.db")

df = pd.read_sql("""
    SELECT DISTINCT td.chembl_id, cs.accession
    FROM target_dictionary td
    JOIN target_components tc ON td.tid = tc.tid
    JOIN component_sequences cs ON tc.component_id = cs.component_id
    WHERE td.organism = 'Homo sapiens'
      AND td.target_type = 'SINGLE PROTEIN'
""", conn)

df.to_csv("data/processed/step3/target_to_uniprot.csv", index=False)
conn.close()
print("âœ… å·²ç”Ÿæˆ data/processed/step3/target_to_uniprot.csv")
```

```python
âœ… æ ¸å¿ƒä»»åŠ¡
å°† UniProt è›‹ç™½æ¡ç›®ï¼ˆAccessionï¼‰ â†’ æ˜ å°„åˆ°æ ‡å‡†äººç±»åŸºå› ç¬¦å·ï¼ˆå¦‚ ABCA3, GATA4ï¼‰

ğŸ”§ å…·ä½“æ“ä½œ
è¯»å–è§£å‹åçš„ UP000005640_9606.datï¼ˆUniProt äººç±»è›‹ç™½å…¨é›†ï¼‰
è§£ææ¯æ¡è›‹ç™½è®°å½•ï¼š
ç”¨ AC è¡Œè·å– UniProt Accessionï¼ˆå¦‚ A6NEH6ï¼‰
ç”¨ GN è¡Œæå– Name=XXX ä¸­çš„åŸºå› åï¼ˆå¦‚ TMEM247ï¼‰
æ¸…ç†æ ¼å¼ï¼š
è½¬å¤§å†™ï¼ˆtmem247 â†’ TMEM247ï¼‰
å»æ‰æ³¨é‡Šï¼ˆå¦‚ {ECO:...}ã€Synonyms=...ï¼‰
è¾“å‡ºå¹²å‡€çš„æ˜ å°„è¡¨ï¼š
data/raw/step3/gene_map.csv
æ ¼å¼ï¼šuniprot,gene_symbol
```

```python
import pandas as pd

rows = []
with open("data/raw/step3/UP000005640_9606.dat") as f:
    acc = None
    for line in f:
        if line.startswith("AC   "):
            # æå–ç¬¬ä¸€ä¸ª Accessionï¼ˆå»æ‰åˆ†å·ï¼‰
            acc = line.split()[1].rstrip(";")
        elif line.startswith("GN   ") and acc:
            # åªå– Name= åçš„å†…å®¹ï¼Œå¿½ç•¥ {ECO:...} å’Œ Synonyms
            gn_part = line[5:].split(";")[0]  # å–ç¬¬ä¸€ä¸ªå­—æ®µï¼ˆName=...ï¼‰
            if "Name=" in gn_part:
                gene = gn_part.split("Name=")[1].strip().split()[0]  # å»æ‰åé¢å¯èƒ½çš„ {ECO...}
                rows.append({"uniprot": acc, "gene_symbol": gene.upper()})
            acc = None  # é˜²æ­¢é‡å¤

# ä¿å­˜
df = pd.DataFrame(rows)
df.to_csv("data/raw/step3/gene_map.csv", index=False)
print(f"âœ… æå– {len(df)} æ¡æ˜ å°„ï¼Œç¤ºä¾‹:")
print(df.head())
```

```python
# 5_find_druggable_targets.py
import pandas as pd
import os

# è·¯å¾„é…ç½®
CHEMBL_GENE_PATH = "data/processed/step3/chembl_target_id_to_gene.csv"
OUTPUT_DIR = "results/step3_outputs/"

# ç¡®ä¿è¾“å‡ºç›®å½•å­˜åœ¨
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 1. è¯»å– ChEMBL é¶ç‚¹-åŸºå› æ˜ å°„
print("ğŸ“¥ æ­£åœ¨åŠ è½½ ChEMBL é¶ç‚¹-åŸºå› æ˜ å°„...")
chembl_gene = pd.read_csv(CHEMBL_GENE_PATH)

# æ£€æŸ¥åˆ—æ˜¯å¦å­˜åœ¨
if "gene_symbol" not in chembl_gene.columns:
    raise KeyError("âŒ 'gene_symbol' åˆ—ä¸å­˜åœ¨ï¼Œè¯·ç¡®è®¤è¾“å…¥æ–‡ä»¶æ ¼å¼")

# 2. è¯»å–é«˜ä¼˜å…ˆçº§é¶ç‚¹ï¼ˆè‚º & å¿ƒè„ï¼‰
print("ğŸ” æ­£åœ¨åŠ è½½é«˜ä¼˜å…ˆçº§é¶ç‚¹åˆ—è¡¨...")
lung_targets = set(
    pd.read_csv(os.path.join(OUTPUT_DIR, "lung_high_priority_targets.csv"))["Gene Symbol"]
)
heart_targets = set(
    pd.read_csv(os.path.join(OUTPUT_DIR, "heart_high_priority_targets.csv"))["Gene Symbol"]
)

# 3. ç­›é€‰å¯è¯é¶ç‚¹
print("ğŸ§ª æ­£åœ¨ç­›é€‰å¯è¯é¶ç‚¹...")
lung_druggable = chembl_gene[chembl_gene["gene_symbol"].isin(lung_targets)].copy()
heart_druggable = chembl_gene[chembl_gene["gene_symbol"].isin(heart_targets)].copy()

# 4. ä¿å­˜ç»“æœ
lung_output = os.path.join(OUTPUT_DIR, "lung_druggable_targets_with_chembl.csv")
heart_output = os.path.join(OUTPUT_DIR, "heart_druggable_targets_with_chembl.csv")

lung_druggable.to_csv(lung_output, index=False)
heart_druggable.to_csv(heart_output, index=False)

# 5. æ‰“å°ç»Ÿè®¡ä¿¡æ¯
lung_unique_chembl = lung_druggable["chembl_id"].nunique()
heart_unique_chembl = heart_druggable["chembl_id"].nunique()

print("\nâœ… ç­›é€‰å®Œæˆï¼")
print(f"ğŸ’Š è‚ºå¯è¯é¶ç‚¹: {len(lung_druggable)} æ¡è®°å½• â†’ å¯¹åº” {lung_unique_chembl} ä¸ªå”¯ä¸€ ChEMBL é¶ç‚¹")
print(f"ğŸ’Š å¿ƒè„å¯è¯é¶ç‚¹: {len(heart_druggable)} æ¡è®°å½• â†’ å¯¹åº” {heart_unique_chembl} ä¸ªå”¯ä¸€ ChEMBL é¶ç‚¹")
print(f"\nğŸ“ ç»“æœå·²ä¿å­˜è‡³:")
print(f"   - {lung_output}")
print(f"   - {heart_output}")
```

```python
# 6_get_drugs_for_targets.py
import pandas as pd
from chembl_webresource_client.new_client import new_client

# 1. åŠ è½½é¶ç‚¹ï¼ˆå« gene_symbolï¼‰
lung = pd.read_csv("results/step3_outputs/lung_druggable_targets_with_chembl.csv")
heart = pd.read_csv("results/step3_outputs/heart_druggable_targets_with_chembl.csv")
all_targets_df = pd.concat([lung, heart]).drop_duplicates(subset=["chembl_id"])
target_ids = all_targets_df["chembl_id"].tolist()

print(f"ğŸ” æ­£åœ¨æŸ¥è¯¢ {len(target_ids)} ä¸ª ChEMBL é¶ç‚¹çš„æ´»æ€§åŒ–åˆç‰©...")

# 2. åˆå§‹åŒ–å®¢æˆ·ç«¯
activities = new_client.activity
molecules = new_client.molecule

# 3. æŸ¥è¯¢
results = []
for target in target_ids:
    print(f"  â†’ æŸ¥è¯¢ {target}...")
    try:
        acts = activities.filter(
            target_chembl_id=target,
            standard_type__in=["IC50", "Ki", "EC50"],
            standard_units="nM",
            standard_value__lte=1000  # â‰¤1 Î¼M
        ).only("molecule_chembl_id", "standard_value", "standard_type")

        for act in acts:
            mol = molecules.get(act["molecule_chembl_id"])
            if not mol or not mol.get("pref_name"):
                continue
            results.append({
                "target_chembl_id": target,
                "gene_symbol": all_targets_df.set_index("chembl_id").loc[target, "gene_symbol"],
                "compound_chembl_id": act["molecule_chembl_id"],
                "compound_name": mol["pref_name"],
                "standard_value": float(act["standard_value"]),
                "standard_type": act["standard_type"],
                "is_approved_drug": mol.get("max_phase") == 4
            })
    except Exception as e:
        print(f"    âš ï¸ è·³è¿‡ {target}: {e}")

# 4. ä¿å­˜
df = pd.DataFrame(results)
output_file = "results/step3_outputs/druggable_targets_compounds.csv"
df.to_csv(output_file, index=False)

approved_count = df["is_approved_drug"].sum()
print(f"\nâœ… å®Œæˆï¼å…±æ‰¾åˆ° {len(df)} ä¸ªæ´»æ€§åŒ–åˆç‰©ï¼Œå…¶ä¸­ {approved_count} ä¸ªæ˜¯ FDA æ‰¹å‡†è¯ç‰©")
print(f"ğŸ“ ç»“æœå·²ä¿å­˜è‡³: {output_file}")
```
